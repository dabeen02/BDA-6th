- 모델에게 학습을 하기 위해서는 "데이터"를 2가지 또는 3가지로 나눈다.
- 왜 데이터를 나누는가?
- 머신에게 모든 데이터를 다 주면 그 전체 데이터의 특성을 다 배우는 것이기 때문에 신뢰할 수 없다.
- 데이터를 나눠서 준다는 개념이 모델이 실제로 학습했을 때 성능이 얼마나 좋은지를 평가하기 위해서이다.
### train, test
- train은 모델에게 모의고사처럼 학습을 시키는 것
- test는 모델을 평가하는 것
- test데이터는 실제로 머신은 본 적이 없고, 어떤 패턴인지 모른다. (학습한 적이 없다)
- train으로 학습한 것을 test를 이용해서 실제 pred값을 만들고, 그 예측 값과 실제 값을 비교해서 성능을 평가한다.

## train, test 나누기

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.2, random_state=111) 
```

## ML 알고리즘으로 타이타닉 생존자 예측해보기
- 의사결정나무
- 랜덤포레스트
- 로지스틱회귀

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# 결정트리, 랜덤포레스트, 로지스틱회귀 만들기
dt_clf = DecisionTreeClassifier(random_state=111)
rf_clf = RandomForestClassifier(random_state=111)
lr_clf = LogisticRegression(solver = 'liblinear')
```

```python
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
print('DT정확도 : {0:.4f}' .format(accuracy_score(y_test, dt_pred)))

rf_clf.fit(X_train, y_train)
rf_pred =rf_clf.predict(X_test)
print('RF정확도 : {0:.4f}' .format(accuracy_score(y_test, rf_pred)))

lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
print('LR정확도 : {0:.4f}' .format(accuracy_score(y_test, lr_pred)))
```
