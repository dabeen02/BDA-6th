## 파생변수 추가하기
- 기존 데이터에 존재했던 변수(mpg기준으로 보면 cty,hwy)에서 새로운 변수(mean_ch)를 추가하는 것
- assign
- cty, hwy의 평균 연비를 만들어보기!
- 평균 연비 만드는 식은 평균 구하는 식과 동일

```python
import numpy as np
```

```python
df = df.assign(mean_ch = (df['cty']+df['hwy'])/2) #mean_ch 새롭게 만들 변수
```

```python
df
```

- np.where 넘파이에서 제공하는 where 함수는 해당 값의 위치를 찾아줌

```python
df.assign(grade = np.where(df['mean_ch']>=20.15, 'Good', 'Bad'))
```

```python
df['mean_ch'].mean()
```

lambda를 이용해서 만들기
- lambda 매개변수 : 표현식
- ex) (lambda x,y: x+y) (10,20) #30

```python
df = df.assign(mean_ch1 = lambda x :(x['cty']+ x['hwy'])/2)
```

```python
df
```

```python
df['mean_ch2'] = (df['hwy']+ df['cty'])/2 #정말 쉽게
```

```python
df
```

```python
np.where(df['mean_ch'] >= 20.15)
```

## Groupby
- 요약 통계를 봐야할 때 수치형 데이터만 사용해서 볼 수 있는데
- 엑셀에서 피벗으로 돌려서 값을 만들 때
- groupby(기준으로 묶을 컬럼), agg(값을 출력할 변수를 넣어주고, 원하는 통계치도 넣어줌) 함수와 같이 사용

```python
df
```

```python
#연도별로 고속도로 연비를 보고싶다
df.groupby('year').agg(mean_cty = ('cty', 'mean'), #('변수', '내가 볼 통계치')
                      sum_cty = ('cty', 'sum'),
                      n = ('model', 'count'))
```

```python
df.groupby('year').agg(mean_cty = ('cty', 'count'), #('변수', '내가 볼 통계치')
                      sum_cty = ('cty', 'mean'),
                      n = ('model', 'sum')) #count는 개수
```

```python
df['model']
```

```python
df.groupby('model').agg(mean_cty = ('cty', 'mean'),
                       sum_cty = ('cty', 'sum'),
                       n = ('model', 'count'))
```

```python
df_2 = df.query("manufacturer == 'audi'").groupby('year').agg(mean_cty = ('cty', 'mean'),
                                                             sum_cty = ('cty', 'sum'),
                                                             n = ('model', 'count'))
```

```python
df_2 #year는 인덱스!
```

```python
df_2.columns #컬럼 3가지: mean_cty, sum_cty, n
```

```python
df_2
```

## (데이터 합치기) merge, concat
- 데이터프레임과 데이터프레임을 합침
- 왜?
- 데이터가 예쁘게 mpg처럼 있는 게 아니라 저장한 공간에 데이터가 테이블 별로 나눠져 있을 건데,
- 그 테이블에 있는 것들을 하나의 mpg 데이터처럼 만들려면 합쳐야 함
- 특정 상황에서 -> 분석가가 분석에 필요한 데이터를 합쳐야 하는 경우에 사용하면 됨
