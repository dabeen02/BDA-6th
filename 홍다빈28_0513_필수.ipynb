## 반복문을 동일한 mpg 값이 아닌 변경하여 배운 반복문 코드를 응용하여 최소 5개 이상 코드 작성하기

```python
import pandas as pd
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg
```

```python
#1 iterrows를 이용하여 displ 정보 추출하기
for idx, row in mpg.iterrows():
    print(idx, row['model'])
```

```python
#2 itertuples를 이용하여 year 정보 추출하기
for row in mpg.itertuples():
    print(row.year)
```

```python
#3 model, year 데이터. model이 a4이고, year가 2008인 데이터 추출
for i, j in zip(mpg.model, mpg.year):
	if i == 'a4' and j== '2008' : 
		print( i , j ) 
```

```python
#4 trans 변수 15인 데이터 추출 후 연도별로 보기
def get_trans_spe(series): 
	tr_cnt = len([ _ for x in series if x[0] in 'l5'])   
	return tr_cnt   

mpg.groupby('year').agg({'trans': get_trans_spe}).reset_index()
```

```python
#5 manufacturer가 volkswagen, year가 2008인 데이터만 추출
manu = ['volkswagen']
year = [2008]

for i , j in zip(mpg.manufacturer, mpg.year):
	if i in manu and j in year : 
		print(i, j) 
```
