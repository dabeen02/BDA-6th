### 상대도수란?
- 도수분포표에서 전체 도수에 대한 어떤 계급의 도수의 비율

### 누적도수란?
- 도수분포표에서 가장 작거나 큰 계급에서 몇 번째 계급까지의 도수의 합

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('scores_em.csv')
df
```

### 상대도수

```python
df_en = df['english']
freq, _ = np.histogram(df_en, bins=10, range=(0,100))

# 데이터프레임으로 만들기
freq_class = [f'{i}-{i+10}' for i in range(0,100,10)]

#데이터프레임 만들기
freq_df = pd.DataFrame({'frequency':freq},
                      index = pd.Index(freq_class, name='class'))

freq_df
```

```python
# 구간별 값을 50(학생 수 총합)으로 나눔
score_freq = freq_df/50
score_freq
```

```python
# check
0.04+0.16+0.32+0.36+0.12
```

### 누적도수

```python
accrue_freq = np.cumsum(score_freq)
accrue_freq
```
