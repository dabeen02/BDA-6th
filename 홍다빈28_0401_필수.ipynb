### 추가 전처리
- 범주화 시키기

```python
# age 범주화
# 우선, 평균과 최대, 최소, 그리고 median 값을 파악
df['age'].mean()
```

```python
def Age(age):
    if age <= 10 : age = 0
    elif age <= 20: age = 1
    elif age <= 30: age = 2
    elif age <= 40: age = 3
    elif age <= 50: age = 4
    elif age <= 60: age = 5
    elif age <= 70: age = 6
    else: age = 7
    
    return age
```

```python
df['age'] = df['age'].apply(lambda x:Age(x))
```

```python
# fare 범주화
# 마찬가지로 평균과 최대, 최소, 그리고 median 값을 파악
df['fare'].mean()
```

```python
df.groupby('fare').agg(n = ('fare', 'count'))
```

```python
def Fare(fare):
    if fare <= 10 : fare = 0
    elif fare <= 50: fare = 1
    elif fare <= 100: fare = 2
    elif fare <= 150: fare = 3
    elif fare <= 200: fare = 4
    elif fare <= 250: fare = 5
    elif fare <= 300: fare = 6
    elif fare <= 350: fare = 7
    elif fare <= 400: fare = 8
    elif fare <= 450: fare = 9
    elif fare <= 500: fare = 10
    else: fare = 11
    
    return fare
```

```python
df['fare'] = df['fare'].apply(lambda x:Fare(x))
```

```python
df = df.select_dtypes(include = [np.number]) #수치형 데이터만 추출하기
```

```python
df= df[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'sex_female', 'sex_male',
       'embarked_C', 'embarked_Q', 'embarked_S', 'class_First', 'class_Second', 'class_Third',
       'who_child', 'who_man', 'who_woman', 'adult_male_False', 'adult_male_True', 'deck_A',
        'deck_B', 'deck_C', 'deck_D', 'deck_E', 'deck_F', 'deck_G', 'embark_town_Cherbourg', 
        'embark_town_Queenstown','embark_town_Southampton', 'alive_no', 'alive_yes', 'alone_False',
       'alone_True' ]]
```

```python
df
```

```python
x_df = df.drop('survived', axis=1)
y_df = df['survived']
```

```python
x_df
```

```python
y_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.2, random_state=111)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dt_clf = DecisionTreeClassifier(random_state=111)
rf_clf = RandomForestClassifier(random_state=111)
lr_clf = LogisticRegression(solver = 'liblinear')
```

```python
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
print('DT정확도 : {0: .4f}' .format(accuracy_score(y_test, dt_pred)))

rf_clf.fit(X_train, y_train)
rf_pred = rf_clf.predict(X_test)
print('RF정확도 : {0: .4f}' .format(accuracy_score(y_test, rf_pred)))

lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
print('LR정확도 : {0: .4f}' .format(accuracy_score(y_test, lr_pred)))
```
