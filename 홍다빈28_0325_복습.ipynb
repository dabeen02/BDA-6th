# 결측치에 대한 접근 방식
- 결측치란 누락값, 비어있는 값을 말한다
- 결측치는 컬럼과 데이터의 관계에 대해서 잘 이해하고 살펴야 한다
- 결측치를 대처하는 자세는?
- 2가지
  - 삭제: 누락값을 지운다 -> 언제, 어떤 경우? 기준이 필요
  - 대체: 다른 값으로 대체하여 채운다 -> 어떤 값으로, 왜, 어떻게? 기준이 필요
- 이 2가지의 답은 해당 데이터 컬럼과 데이터의 관계에 따라 다를 수 있고
- 데이터 분석가의 도메인 지식이 정말 중요하다

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'이름': ['홍길동', '김영희', '배철수'],
                  '성적': [50, np.nan, 30],
                  '등급': ['A', 'B', np.nan]})
```

```python
df
```

```python
# na값 결측치 missing value는 어떻게 보는가?
df.isna()
```

```python
df.isna().sum()
```

1. 삭제
- df.dropna(subset=['컬럼'])

```python
df.dropna(subset=['성적'])
```

```python
df.dropna() #지금은 컬럼이 1개만 na 값이라 동일한 결과지만 컬럼이 많으면 내가 날리고 싶지 않은 것도 다 날려진다.
```

```python
df.dropna(subset=['성적', '등급'])
```

- 날리는 기준? 정해진 건 없다
- 작은 비중의 결측치가 존재하면 최대한 대체할 수 있는 부분을 대체하자
- 결측치 비중이 너무 적어서 분석에 큰 중요도가 없으면 삭제할 수도 있다 

```python
df
```

- 성적에 대해서 na값을 대체하기!
- 성적에 있는 na값을 대체 값으로 넣어야 한다

2. 대체
- fillna()

```python
df['성적']= df['성적'].fillna(100)
```

```python
df_mean = df.성적.mean()
df['성적'] = df['성적'].fillna(df_mean)
```

```python
df['성적']= df['성적'].fillna((50+30)/2)
```

```python
df
```

- na값에 대한 통계치 계산
- 통계치를 구할 때 na값은 반영이 안 된다 (예: 평균 구할때)
- 값의 개수를 셀 때 na값도 반영이 안 된다

- 연령은 10, 20, 30, 40층을 나눠서 비교해도 충분히 이해가 될 수 있다
- 수치형 데이터는 범주화 시켜서 하나의 범주형 카테고리로 연령을 만들어도 된다

```python
import matplotlib.pyplot as plt
```

```python
group = ['baby', 'child', 'student', 'Young adult', 'Adult', 'Elderly']
```

```python
group
```

```python
def cat_age(age):
    cat = ''
    if age <=5 :cat ='baby'
    elif age <=18: cat='child'
    elif age <=30: cat='student'
    elif age <=40: cat='Young adult'
    elif age <=60: cat='Adult'
    else: cat ='Elderly'
    
    return cat
```

```python
plt.figure(figsize=(15,10))
df['cat_age'] = df['age'].apply(lambda x:cat_age(x))
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='cat_age',y='survived',hue='sex', data = df, order = group)
```
