- 타이타닉 데이터로 수업 시간에 머신러닝을 진행할 건데 여러분들이 전처리한 데이터를 가지고 진행할 것
- 여러분들이 전처리한 데이터에 대한 정확도나 다른 지표들을 확인 가능
- 전처리는 모두 동일하게 하지 않을 것이기 때문에 실제 다른 분들의 전처리를 보고 그걸 통해 나온 정확도 등을 확인해서 어떤지 알아보는 시간

### 타이타닉 데이터 전처리 진행
### 문자열 데이터가 다 인코딩 되고, 수치형 데이터도 결측치가 없는 하나의 깔끔한 데이터 프레임으로 만들어 제출
### 파일 형식은 csv 파일 + 구체적인 코드와 주석과 왜 전처리를 그렇게 했는지 정확하게 설명하여 제출

```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum() #결측치 확인
```

- 결측치: age(수치형), deck(문자열), embark_town(문자열)
- age: 나이
- deck: 선실 번호 첫 알파벳
- embarked / embark_town: 탑승지 이름

- 결측치 대체: fillna()

- age 대체
- 평균보다 중앙값으로 대체하는 것이 좀 더 안정적이므로 중앙값으로 대체

```python
age_median = df.age.median() #중앙값 계산
```

```python
age_median
```

```python
df['age'] = df['age'].fillna(age_median) #중앙값 대체
```

```python
df
```

```python
df.isna().sum() #age 결측값 대체 확인 완료
```

- embarked / embark 대체

```python
df.embarked.value_counts() #각 선실 번호 개수 확인
```

- Q가 제일 적으므로 embarked/embark_town의 결측치 값을 모두 Q(Queenstown)으로 대체

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].min)
```

```python
df
```

```python
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].min)
```

```python
df
```

```python
df.isna().sum() #embarked/embark_town 결측값 대체 확인
```

- deck 대체

```python
df.deck.value_counts() #각 deck 개수 확인
```

- C가 가장 많으므로 C로 대체 (가장 단순한 대체)

```python
df['deck'] = df['deck'].fillna('C')
```

```python
df
```

```python
df.isna().sum() #deck 결측값 대체 확인
```

### 모든 결측값 대체 완료!

## 인코딩

- 인코딩할 변수 찾기

```python
df.info()
```

- 인코딩할 변수: sex, embarked, class, who, adult_male, deck, embark_town, alive, alone
- 원핫인코딩으로 진행
- 원핫인코딩은 0과 1로 표현하는 것
- 판다스에서 제공해 주는 함수 get_dummies를 사용하여 쉽게 원핫인코딩 0과 1로 바꿀 수 있다

```python
#sex
df = pd.get_dummies(df, columns=['sex'])
```

```python
df
```

- 컬럼이 sex 대신 sex_female과 sex_male로 대체되고, 원핫인코딩 0,1로 표현된 것을 확인

```python
#embarked
df = pd.get_dummies(df, columns=['embarked'])
```

```python
df
```

- 컬럼이 embarked 대신 embarked_C와 embarked_Q, embarked_S로 대체되고, 원핫인코딩 0,1로 표현된 것을 확인
- 컬럼 이름이 embarked의 각 변수 값으로 대체됨

```python
#class
df = pd.get_dummies(df, columns=['class'])
```

```python
df
```

- 컬럼이 class 대신 class_First와 class_Second, class_Third로 대체되고, 원핫인코딩 0,1로 표현된 것을 확인

```python
#who
df = pd.get_dummies(df, columns=['who'])
```

```python
df
```

- 컬럼이 who 대신 who_child와 who_man, who_woman로 대체되고, 원핫인코딩 0,1로 표현된 것을 확인

```python
#adult_male
df = pd.get_dummies(df, columns=['adult_male'])
```

```python
df
```

- 컬럼이 who 대신 adult_male_False와 adult_male_True로 대체되고, 원핫인코딩 0,1로 표현된 것을 확인

```python
#deck
df = pd.get_dummies(df, columns=['deck'])
```

```python
df
```

- 컬럼이 deck 대신 각 변수인 A부터 G로 대체되고, 원핫인코딩 0,1로 표현된 것을 확인

```python
#embark_town
df = pd.get_dummies(df, columns=['embark_town'])
```

```python
df
```

- 컬럼이 embark_town 대신 각 변수인 Cherbourg, Queenstown, Southampton로 대체되고, 원핫인코딩 0,1로 표현된 것을 확인

```python
#alive
df = pd.get_dummies(df, columns=['alive'])
```

```python
df
```

- 컬럼이 alive 대신 alive_no, alive_yes로 대체되고, 원핫인코딩 0,1로 표현된 것을 확인

```python
#alone
df = pd.get_dummies(df, columns=['alone'])
```

```python
df
```

- 컬럼이 alone 대신 alone_False, alone_True로 대체되고, 원핫인코딩 0,1로 표현된 것을 확인

```python
df.isna().sum()
```
