# 타이타닉 데이터 컬럼 설명
- survived: 생존 여부 / 0이면 사망, 1이면 생존
- pclass: 객실 등급 / 1이면 1등급, 2이면 2등급, 3이면 3등급
- sex: 성별 / male이면 남자, female이면 여자
- age: 연령
- sibsp: 타이타닉호에 탑승한 형제 및 배우자의 수
- parch: 타이타닉호에 탑승한 부모 및 자녀의 수
- fare: 승객 요금
- embarked: 탑승지 이름 앞글자 / C(Cherbourg), Q(Queenstown), S(Southampton)
- class: 객실 등급 / First면 1등급, Second면 2등급, Third면 3등급
- who:남자, 여자, 아이 / man, woman, child
- adult_male: 성인 남자인지 여부 / True면 성인 남자, False면 그외
- deck: 선실 번호 첫 알파벳 / A,B,C,D,E,F,G
- embark_town: 탑승지 이름 / Cherbourg, Queenstown, Southampton
- alive: 생존 여부 / no면 사망, yes면 생존
- alone: 혼자 탑승했는지 여부 / True면 혼자 탑승, False면 가족과 함께 탑승

# 질문 10가지

1. 생존자/미생존자 각 인원은?
2. 생존자 중 가장 많이 탑승한 곳은?
3. 생존자/미생존자의 평균 연령을 제시하시오
4. 생존자 중 미성년자를 정렬하시오 (오름차순으로)
5. 남녀 수를 생존자별로 제시하시오
6. 생존자들 객실 등급의 각 개수를 구하시오
7. 생존자들의 선실 번호를 정렬하시오 (오름차순으로)
8. 형제 및 배우자 그리고 부모 및 자녀와 함께 탑승한 생존자들의 인원을 각각 제시하시오
9. 20대부터 40대까지의 생존자 중 혼자 오신 분의 인원은?
10. 생존한 미성년자를 'yes'로, 아니면 'no'로 표현하여 제시하시오 (새로운 변수 이름: survived_child)

-------------------------------------------------------------------------------------------------------------------------
1. 생존자/미생존자 각 인원은?

```python
df.groupby('survived').agg(n = ('survived', 'count'))
```

생존자: 342명, 미생존자: 549명

-------------------------------------------------------------------------------------------------------------------------
2. 생존자 중 가장 많이 탑승한 곳은?

```python
df.query("survived == 1").groupby('embark_town').agg(n = ('embark_town', 'count'))
```

Southampton

-------------------------------------------------------------------------------------------------------------------------
3. 생존자/미생존자의 평균 연령을 제시하시오

```python
df.groupby('survived').agg(mean_age = ('age', 'mean'))
```

-------------------------------------------------------------------------------------------------------------------------
4. 생존자 중 미성년자를 정렬하시오 (오름차순으로)

```python
df.query("survived == 1 and age <= 19")[['age']].sort_values('age')
```

-------------------------------------------------------------------------------------------------------------------------
5. 남녀 수를 생존자별로 제시하시오

```python
df.groupby(['survived', 'sex']).agg(n=('survived', 'count'))
```

-------------------------------------------------------------------------------------------------------------------------
6. 생존자들 객실 등급의 각 개수를 구하시오

```python
df.query("survived == 1").groupby('class').agg(n = ('class', 'count'))
```

First: 136개, Second: 87개, Third: 119개

-------------------------------------------------------------------------------------------------------------------------
7. 생존자들의 선실 번호를 정렬하시오 (오름차순으로)

```python
df.query("survived == 1")[['deck']].sort_values('deck')
```

-------------------------------------------------------------------------------------------------------------------------
8. 형제 및 배우자 그리고 부모 및 자녀와 함께 탑승한 생존자들의 인원을 각각 제시하시오

```python
df.query("survived==1 and sibsp >= 1 and parch >= 1").groupby(['sibsp', 'parch']).agg(n_sibsp = ('sibsp', 'count'),
                                                                                     n_parch = ('parch', 'count'))
```

-------------------------------------------------------------------------------------------------------------------------
9. 20대부터 40대까지의 생존자 중 혼자 오신 분의 인원은?

```python
df.query("age >= 20 and age < 50 and survived == 1").groupby('alone').agg(n = ('alone', 'count'))
```

100명

-------------------------------------------------------------------------------------------------------------------------
10. 생존한 미성년자를 'yes'로, 아니면 'no'로 표현하여 제시하시오 (새로운 변수 이름: survived_child)

```python
import numpy as np
df.assign(survived_child = np.where((df['survived'] == 1) & (df['age']<=19), 'yes', 'no'))
```
